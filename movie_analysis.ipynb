```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
movies = pd.read_csv('movies.csv')
import re
```

## What was the runtime for each movie?

The graph below shows the runtime for each movie, from longest to shortest.

```python
runtimes = movies.sort_values(by='Runtime', ascending=False)
runtimes.plot.bar(x='Movie Title', y='Runtime', xlabel='Movie', ylabel='Runtime (mins)', title='Movies by Runtime', color='#84511d')
```

## What movie had the longest runtime?

The table below shows all movies in the CSV file, listed from longest to shortest.

```python
movies.sort_values(by='Runtime', ascending=False)
```

## Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?

The scatter plot below shows the relationship between the runtime of a movie and the number of awards a movie won.

```python
length_award_plot = movies.sort_values(by='Runtime', ascending=False)
length_award_plot.plot.scatter(x='Runtime', y='Award Wins', title='Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?', color='#35a4ca')
```

As we can see, there is no clear correlation between a movie's runtime and the respective number of awards they won.

#### Hence, there is no clear relationship between a movie's runtime and the respective number of awards won.

## How many awards and nominations did each movie receive?

The tables below respectively show all movies listed in order of number of awards and nominations won, from most to least.

```python
movies.sort_values(by='Award Wins', ascending=False)
```

```python
movies.sort_values(by='Award Nominations', ascending=False)
```

The multiple bar chart below shows the number of awards and nominations each movie in the dataset received.

```python
x = movies['Movie Title']
awards = movies['Award Wins']
nominations = movies['Award Nominations']

x_axis = np.arange(len(x))

df = pd.DataFrame({'Movie Title': x_axis, 'Award Wins': awards, 'Award Nominations': nominations})
df_sorted = df.sort_values('Award Wins', ascending=False)

plt.bar(x, awards, color='#22dd6b')
plt.bar(x, nominations, bottom=awards, color='#e7184e')
plt.xticks(x_axis, x, rotation=90)
plt.xlabel('Movie Title')
plt.ylabel('Number of Awards/Nominations Received')
plt.title('How many awards and nominations did each movie receive?')
plt.legend(['Award Wins', 'Award Nominations'])
plt.show()
```

## Is there a relationship between the amount of box office earnings a movie had and the amount of total nominations the movie received (total nominations = awards wins + award nominations)?

The scatter plot below shows the relationship between box office earnings and total nominations.

```python
box_office_plot = movies.sort_values(by='Box Office', ascending=False)
x = np.array(movies['Box Office'])
y = np.array(movies['Nominations'])
plt.scatter(x, y, color='#be1ae5')
plt.xticks(rotation=90)
plt.xlabel('Box Office Earnings')
plt.ylabel('Total Nominations')
plt.title('Relationship Between Box Office Earnings and Total Nominations')
plt.ticklabel_format(style='plain')
plt.show()
```

#### As there is no correlation between the two variables, hence, there is no clear relationship between box office earnings and total number of nominations.

## Is there a relationship between box office earnings and movie runtimes?

```python
box_run_plot = movies.sort_values(by='Box Office', ascending=False)
x = np.array(movies['Box Office'])
y = np.array(movies['Runtime'])
plt.scatter(x, y, color='#beb541')
plt.xticks(rotation=90)
plt.xlabel('Box Office Earning')
plt.ylabel('Movie Runtime')
plt.title('Relationship Between Box Office Earnings and Movie Runtimes')
plt.ticklabel_format(style='plain')
plt.show()
```

#### There is a weak positive correlation between the two variables, hence there is a slightly positive relationship between box office earnings and movie runtimes.

## How much did each movie earn at the box office?

The table below shows the box office performance for each movie in the data set, from highest grossing to lowest grossing.

```python
movies.sort_values(by='Box Office', ascending=False)
```

The bar chart below shows the same information presented above.

```python
x_axis = movies['Movie Title']
box_office = movies['Box Office']
df = pd.DataFrame({"Movie Title":x_axis, "Box Office":box_office})
df_sorted = df.sort_values('Box Office', ascending=False)

plt.bar('Movie Title', 'Box Office', color='#f10eb5', data=df_sorted)
plt.xticks(x_axis, rotation=90)
plt.xlabel('Movie Title')
plt.ylabel('Box Office')
plt.title('How much did each movie earn at the box office?')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.show()
```

## What is the total count of each genre present in the dataset?

The table below shows the different genres of each movie in the dataset.

```python
print(movies["Genre"])
```

Action, Adventure, Biography, Comedy, Crime, Drama, Fantasy, History, Mystery, Romance, Sci-Fi, Thriller are the different genres mentioned above.

The bar chart below shows the total count of genres present in the dataset.

```python
action_pattern = r'Action'
action = movies["Genre"].str.contains(action_pattern).sum()

adventure_pattern = r'Adventure'
adventure = movies["Genre"].str.contains(adventure_pattern).sum()

biography_pattern = r'Biography'
biography = movies["Genre"].str.contains(biography_pattern).sum()

comedy_pattern = r'Comedy'
comedy = movies["Genre"].str.contains(comedy_pattern).sum()

crime_pattern = r'Crime'
crime = movies["Genre"].str.contains(crime_pattern).sum()

drama_pattern = r'Drama'
drama = movies["Genre"].str.contains(drama_pattern).sum()

fantasy_pattern = r'Fantasy'
fantasy = movies["Genre"].str.contains(fantasy_pattern).sum()

history_pattern = r'History'
history = movies["Genre"].str.contains(history_pattern).sum()

mystery_pattern = r'Mystery'
mystery = movies["Genre"].str.contains(mystery_pattern).sum()

romance_pattern = r'Romance'
romance = movies["Genre"].str.contains(romance_pattern).sum()

sci_fi_pattern = r'Sci-Fi'
sci_fi = movies["Genre"].str.contains(sci_fi_pattern).sum()

thriller_pattern = r'Thriller'
thriller = movies["Genre"].str.contains(thriller_pattern).sum()

genres_axis_labels = ["Action", "Adventure", "Biography", "Comedy", "Crime", "Drama", "Fantasy", "History", "Mystery", "Romance", "Sci-Fi", "Thriller"]
all_genres = [action, adventure, biography, comedy, crime, drama, fantasy, history, mystery, romance, sci_fi, thriller]

df_raw = pd.DataFrame({'Genre Type': genres_axis_labels, 'Occurrences': all_genres})
df = df_raw.sort_values(by='Occurrences', ascending=False)
df.plot.bar(x='Genre Type', y='Occurrences', xlabel='Genre', ylabel='Number of occurrences', title='What is the total count of each genre present in the dataset?', color='#d9264c')
```

#### Drama is the most common genre amongst all the movies in the dataset (13), crime, fantasy, mystery and sci-fi each being the least, with only 1 occurrence.
